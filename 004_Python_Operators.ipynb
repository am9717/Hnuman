# Python Operators

Operators are special symbols that perform specific operations on one or more operands (values) and then return a result. 

**For example, you can calculate the sum of two numbers using an addition (+) operator.**



**Python has seven types of operators that we can use to perform different operation and produce a result.**

1. Arithmetic operator
2. Assignment operators
3. Relational operators
4. Logical operators
5. Membership operators
6. Identity operators

## 1. Arithmatic Operators

Arithmetic operators are used to perform **mathematical operations** like **addition**, **subtraction**, **multiplication** etc.



#### Example : 

```python
# creating two variables to store 12 and 34

num1 = 12
num2 = 34

# Addition
print("Addition of 12 & 34 is",num1+num2)

# Subtraction
print("Subtraction of 12 & 34 is",34-12)

# Multiplicaion
print("Multiplicaion of 12 & 34 is",12*34)

# Division
print("Division of 12 & 34 is",34/12) # Floating point number

# Integer Division
print("Integer Division of 12 & 34 is",34//12) # Rounds down the result

# Modulus
print("Modulus of 12 & 34 is",12%34) # Gives us remainder

# Power/ Exponent
print("Power of 12 to 4 is",12**4) # Power
```

## 2. Assignment operators

Assignment operators are used in Python to **assign values** to **variables**.



#### Example :

```python
# Assignment Operator
a = 10

# Add & assign
a+=10  # 20 
       # a = a + 10
print(a)

# Subtract & assign
a-=10  # 10
       # a = a - 10
print(a)

# Multiplication & assign
a*=10  # 100
       # a = a * 10
print(a)

# Division & assign
a/=10  # 10
       # a = a / 10
print(a)
```

## 3. Comparison/Relational operators

Comparison operators are used to **compare values**. It either returns **True** or **False** according to the **condition**.



#### Example : 

```python
print(3<5)        # True
print(3>5)        # False
print(3==3)       # True
print(3==3.0)     # True
print(3!=3)       # False
print(12<=7)      # False
print(12>=7)      # True
```

## 4. Logical/Boolean operators 

Logical operators are the **`and`**, **`or`**, **`not`** operators.



#### Example : Logical operators in Python

```python
# Truth table for 'and' operator
# True + True ---->    True
# False + True ---->   False
# True + False ---->   False
# False + False ---->  False

# True + True + False + True ---> False

# Truth table for 'or' operator
# True + True ---->    True
# False + True ---->   True
# True + False ---->   True
# False + False ---->  False

# False + True + False + False ---> True

# Using and operator
print(1<1 and 1!=1)  # False 
print(1<2 and 1!=1)  # False

# Using or operator
print(1<1 or 1!=1)   # False 
print(1<2 or 1!=1)   # True

# Using not operator
print(not 1<1 and not 1!=1) # True
```

### 6. Identity operators

Identity operators are used to check whether the value of two variables is the same or not. This operator is known as a reference-quality operator because the identity operator compares values according to two variables' memory addresses.



#### Example : 

```python
3 is 3.0
```

```python
3 == 3.0
```

```python
id(3)
```

```python
id(3.0)
```

```python
x = 437251555 
y = x

x is y 
```
