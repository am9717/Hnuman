### Programming Paradigm: An approach to organize & structure your code

### There are 3 Paradigms in programming:

### 1. Imperative : Doing programming with diffferent loops, if/else statement and general basics.
### 2. Functional : Doing programming with pure, higher order, recursive and all types of functions.
### 3. Object oriented: Doing programming with classes and objects.

## what is a class?

- A class is an object's blueprint or template.
- It's a data type that's been specified by the user.
- A class defines variables, constants, member functions, and other functionality.



## what is an object?

- An object is basically a collection of properties which can be expressed as variables and some functions.
- We need to put the object in a variable.
- That variable will contain whole set of properties of the object.
- Object contains data which are unique to each other.



## Difference Between Classes And Objects



## What is attribute?

- The variables within the object are called either "Instance variables" or "attributes".

## What is method?

The functions within the object are called methods.

```python
# Defining class with a constructor

class Employee:
     # __init__ 
    def __init__(self,fname,lname,age): # attributes/instance variables
        self.fname=fname
        self.lname=lname
        self.age=age

    # instance method
    def display_info(self):
        print(f"First Name: {self.fname}")
        print(f"Last Name: {self.lname}")
        print(f"Age: {self.age}")

# Creating a new object

employee1 = Employee('James','Webb',23)
employee2 = Employee('Hina','Sharma',45)

# Displaying information

# First Employee Inofrmation
print("First Employee Inofrmation")
employee1.display_info()

print("Second Employee Inofrmation")
employee2.display_info()
```

```python
## Define calculator using OOP

class Calculator:
    def __init__(self,num1,num2):
        self.num1=num1
        self.num2=num2

    def add(self):
        print("Sum:", self.num1+self.num2)

    def subtraction(self):
        print("Subtraction:", self.num1-self.num2)

    def multiplication(self):
        print("Product:", self.num1*self.num2)

    def division(self):
        print("Division:", self.num1/self.num2)

num1 = int(input("Enter number 1: "))
num2 = int(input("Enter number 2: "))

c1 = Calculator(num1,num2)

c1.add()
c1.add()
```

```python
def addition():
    return num1+num2

def subtraction():
    return num1-num2

def multiplication():
    return num1*num2

def division():
    return num1/num2

while True:
    print("""
    1: Addition
    2: Subtraction
    3: Multiplication
    4: Division
    5: Exit...
    """)

    ch = int(input("Select Operation: "))

    if ch in (1,2,3,4,5):
        # first check whether we want to exit
        if ch==5:
            break
    
        # if not, then ask user to enter his/her choice
        num1 = int(input("Enter first number: "))
        num2 = int(input("Enter second number: "))
    
        if ch==1:
            print("Addition:",addition())
        elif ch==2:
            print("Subtraction:",subtraction())
        elif ch==3:
            print("Multiplication:",multiplication())
        elif ch==4:
            print("Division:",division())
    
else:
    print("Input is not valid")
```

```python
class Calc:
    
    def addition(self,num1,num2):
        return num1+num2
    
    def subtraction(self,num1,num2):
        return num1-num2
    
    def multiplication(self,num1,num2):
        return num1*num2
    
    def division(self,num1,num2):
        return num1/num2

# creating an object
myCalc = Calc()

while True:
    print("""
    1: Addition
    2: Subtraction
    3: Multiplication
    4: Division
    5: Exit...
    """)

    ch = int(input("Select Operation: "))

    if ch in (1,2,3,4,5):
        # first check whether we want to exit
        if ch==5:
            break
    
        # if not, then ask user to enter his/her choice
        num1 = int(input("Enter first number: "))
        num2 = int(input("Enter second number: "))
    
        if ch==1:
            print("Addition:",myCalc.addition(num1,num2))
        elif ch==2:
            print("Subtraction:",myCalc.subtraction(num1,num2))
        elif ch==3:
            print("Multiplication:",myCalc.multiplication(num1,num2))
        elif ch==4:
            print("Division:",myCalc.division(num1,num2))
    
else:
    print("Input is not valid")
```
